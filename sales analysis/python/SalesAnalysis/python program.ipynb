# merging all our 12 months sales together as one CSV file

```python
allfile =[file for file in os.listdir('./Sales_Data')]

allmonthdata=pd.DataFrame()

for file in allfile:
    pp=pd.read_csv('./Sales_Data/'+file)
    allmonthdata=pd.concat([allmonthdata,pp])
    allmonthdata.to_csv('newcsv')
    
```

```python
df=pd.read_csv('newcsv')
```

```python
df.sample(10)
```

# droping unwanted column (Unnamed: 0)

```python
df=df.drop(columns='Unnamed: 0',index=1)
```

```python
df
```

# Task 1.. creating month column

```python
df['Month']=df['Order Date'].str[0:2]
```

# Now we will drop all the NAN rows

```python
df
```

# converting month column datatype to int   df['Month']=df['Month'].astype(int)

# it shows an error that month column contains values 'OR'

# so lets look for all the value in month column that contain 'OR' 

```python
df[df['Month']=='Or']
```

# lets remove the row with 'OR'

```python
df=df[df['Month']!='Or']
```

```python
df
```

# now we can convert month datatype to int without getting an error

```python
df['Month']=df['Month'].astype(int)
```

```python
df.dtypes
```

# convert Quantity Ordered column to int

```python
df['Quantity Ordered']=df['Quantity Ordered']=pd.to_numeric(df['Quantity Ordered'])
```

# convert price column to int

```python
df['Price Each']=df['Price Each']=pd.to_numeric(df['Price Each'])
```

# creating sales column

```python
df['Sales']=df['Quantity Ordered']*df['Price Each']
```

```python
df
```

# Task 2 what was the best month for sales and how much was earn that month?

```python
df_sum=df.groupby('Month').sum()
```

```python
df_sum
```

# Lets visualize it

```python
import matplotlib.pyplot as plt
```

```python
month_plot=[m for m, ss in df.groupby('Month')]

plt.figure(figsize=(10,7))

plt.bar(month_plot, df_sum['Sales'],color='green')
plt.grid()
plt.title('what is the best month for sale')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation='vertical')
plt.yticks(df_sum['Sales'])
plt.show()
```

# Task 3... add city column

```python
def new_city(address):
    return address.split(',')[1]
```

```python
df['City']=df['Purchase Address'].apply(lambda x: new_city(x))
```

```python
df
```

# Task 4... what city has the highest number of sales?

```python
city_sum=df.groupby('City').sum()
```

```python
city_sum
```

# lets visualize it

```python
city_plot=[m for m,ss in df.groupby('City')]

plt.figure(figsize=(10,7))


plt.bar(city_plot,city_sum['Sales'],color='r')
plt.title('WHICH CITY HAS THE HIGHEST SALES')
plt.xlabel('CITY')
plt.ylabel('SALES')
plt.yticks(city_sum['Sales'])
plt.xticks(rotation='vertical')
plt.grid()
plt.show()
```
